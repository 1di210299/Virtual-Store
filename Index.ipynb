```python
#iNTEGRANTES: JUAN DIEGO GUTIERREZ (20172174) , JUAN DIEGO LINARES (20202067)
# Paso 1: Creación de Inventario
# Se crea un diccionario llamado inventory con los cinco productos solicitados, además de su precio y cantidad.
inventory = {
    'Producto1': {'precio': 10.99, 'cantidad': 50},
    'Producto2': {'precio': 24.99, 'cantidad': 30},
    'Producto3': {'precio': 8.50, 'cantidad': 20},
    'Producto4': {'precio': 15.75, 'cantidad': 40},
    'Producto5': {'precio': 12.00, 'cantidad': 25},
}
# Se utiliza una variable para registrar las ventas, con el objetivo de que se actualice a la hora de ejecutar el 
# paso 4 cada vez que haya una venta, y este solo muestre el monto de lo que se ha vendido.
ventas_realizadas = {}
```

```python
# Paso 2: Mostrar Inventario y Comprar
# Se define la función display_inventory, para mostrar el inventario de la tienda
def display_inventory():
    # Se imprime un encabezado para indicar que se está mostrando el inventario
    print("Inventario de la tienda:")
    # Se utiliza bucle for que itera sobre el diccionario creado. La variable producto almacena la clave (productos), y detalles 
    # almacena los valores (precio y cantidad)
    for producto, detalles in inventory.items():
        # Se imprime información sobre cada producto, incluyendo su nombre, precio y cantidad en stock mediante una cadena
        # formateada para combinarlos y aparezcan de esa forma.
        print(f"{producto}: Precio - ${detalles['precio']}, Cantidad en stock - {detalles['cantidad']}")

# Se define la función purchase_product, para solicite al usuario el nombre del producto que desea comprar y la cantidad.
def purchase_product():
    # Se hace uso de la función anteriormente definida para mostrar al usuario el inventario antes de su compra.
    display_inventory()
    # Con el input, se solicita al usuario el nombre del producto que desea comprar.
    producto_deseado = input("Ingrese el nombre del producto que desea comprar: ")

    # Es necesario usar un condicional if para corroborar que el producto ingresado por el usuario se encuentra en el inventario
    if producto_deseado in inventory:
        # Si es el caso, entonces se solicita al usuario, por medio del input, que ingrese la cantidad que desea comprar del
        # producto que solicitó.
        cantidad_deseada = int(input("Ingrese la cantidad que desea comprar: "))
        # Es necesario usar otro condicional para corroborar que la cantidad que se solicita está disponible en el inventario.
        # Por ello, la 'cantidad deseada' por el usuario debe ser menor o igual a la cantidad que se encuentra en el inventario.
        if cantidad_deseada <= inventory[producto_deseado]['cantidad']:
            # Se calcula el precio total de la compra multiplicando la cantidad que solicita el usuario por el precio unitario
            # del producto.
            precio_total = cantidad_deseada * inventory[producto_deseado]['precio']
            print(f"El precio total es: ${precio_total}")
            
            # Se verifica si el producto_deseado no está presente como clave en el diccionario ventas_realizadas;
            # si no está presente, significa que no ha habido ventas registradas anteriormente para ese producto.
            if producto_deseado not in ventas_realizadas:
                # Si el producto no ha sido registrado previamente en ventas_realizadas, esta línea establece una entrada 
                # para el producto con un valor inicial de 0. Esto se hace para garantizar que haya una entrada para cada 
                # producto, incluso si aún no se ha vendido.
                ventas_realizadas[producto_deseado] = 0
                # Se suma la cantidad recién vendida (cantidad_deseada) al valor existente en el 
                # diccionario ventas_realizadas. Esto refleja el seguimiento acumulado de la cantidad total vendida 
                # del producto en cuestión.
            ventas_realizadas[producto_deseado] += cantidad_deseada
            
            # Con esto sea ctualiza la cantidad en stock del producto, de tal forma que queda lo que aún no se ha vendido.
            inventory[producto_deseado]['cantidad'] -= cantidad_deseada
        
        # Si no se da el caso de que la cantidad solicitada es menor o igual al inventario, se imprime lo siguiente:
        else:
            print("Lo siento, no hay suficiente cantidad en stock.")
            
    # Si el producto ingresado no se identifica en el inventario, entonces se imprime lo siguiente:
    else:
        print("Producto no disponible en el inventario.")
```

```python
# Paso 3: Generar Informe

# Se define la función generate_sales_report para generar y mostrar informes de venta
def generate_sales_report():
    # Se establece que, inicialmente, el ingreso por las ventas es 0.
    total_ingresos = 0
    print("Informe de ventas:")
    
    # Se utiliza un bucle for que itera sobre todos los productos en el inventario
    for producto, detalles in inventory.items():
        # Se obtiene la cantidad vendida de cada producto, o se obtiene 0 si no se ha vendido nada de algún producto.
        cantidad_vendida = ventas_realizadas.get(producto, 0)
        # Se obtiene el precio de cada producto.
        precio_producto = detalles['precio']
        # Con esto, se logra obtener el total de ingresos de cada producto, como resultado de las ventas del producto en cuestión.
        ingresos_producto = cantidad_vendida * precio_producto
        # Se suma los ingresos por cada producto
        total_ingresos += ingresos_producto
        # Se muestra la cantidad vendida actualizada y el total de ingresos por las ventas de cada producto.
        print(f"{producto}: Cantidad vendida - {cantidad_vendida}, Ingresos - ${ingresos_producto}")
    
    # Se muestra el total de ingresos 
    print(f"Ingreso total de la tienda: ${total_ingresos}")
```

```python
# Paso 4: Menú de Opciones
# Se crea un bucle para el menú de opcines
while True:
    # Se imprime cada opción disponible en el menú de opcines
    print("\nMenú de opciones:")
    print("1. Mostrar inventario")
    print("2. Comprar producto")
    print("3. Generar informe")
    print("4. Salir")
    
    # Se coloca un input para que el usuario elija a que opción quiere acceder
    opcion = input("Ingrese el número de la opción deseada: ")

    # Se evalúa cada opción ingresada por el usuario, y entonces se ejecuta la acción correspondiente en base a los códigos
    # anteriormente establecidos.
    if opcion == '1':
        display_inventory()
    elif opcion == '2':
        purchase_product()
    elif opcion == '3':
        generate_sales_report()
    # Ya que la opción 4 es salir, se imprime un mensaje de despedida al usuario, y se culmina el bucle.
    elif opcion == '4':
        print("Gracias por usar nuestro programa. ¡Hasta luego!")
        break
    
    # Sino se ingresa una de las opciones disponibles en el menú, entonces se muestra lo siguiente: 
    else:
        print("Opción no válida. Por favor, ingrese un número válido.")
```
